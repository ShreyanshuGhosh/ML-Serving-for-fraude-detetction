DATA PREPROCESSING

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
train_df = pd.read_csv("..\data\FraudTrain.csv")
test_df = pd.read_csv("..\data\FraudTest.csv")

#newtest  = test

```

```python
cols_to_drop = [
'Unnamed: 0','cc_num','first','last','street','trans_num'
]

train_df = train_df.drop(columns=cols_to_drop)
test_df = test_df.drop(columns=cols_to_drop)
```

```python
# change date and time 
train_df['trans_date_trans_time'] = pd.to_datetime(train_df['trans_date_trans_time'])
test_df['trans_date_trans_time'] = pd.to_datetime(test_df['trans_date_trans_time'])

train_df['hour'] = train_df['trans_date_trans_time'].dt.hour
test_df['hour'] = test_df['trans_date_trans_time'].dt.hour

train_df = train_df.drop(columns=['trans_date_trans_time'])
test_df = test_df.drop(columns=['trans_date_trans_time'])



# change dob
train_df['dob'] = pd.to_datetime(train_df['dob'])
test_df['dob'] = pd.to_datetime(test_df['dob'])

train_df['age'] = 2026 - train_df['dob'].dt.year
test_df['age'] = 2026- test_df['dob'].dt.year

train_df = train_df.drop(columns=['dob'])
test_df = test_df.drop(columns=['dob'])



#Distance Feature (generally long distance )
train_df['distance'] = np.sqrt(
(train_df['lat'] - train_df['merch_lat'])**2 +
(train_df['long'] - train_df['merch_long'])**2
)

test_df['distance'] = np.sqrt(
(test_df['lat'] - test_df['merch_lat'])**2 +
(test_df['long'] - test_df['merch_long'])**2
)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
cat_col = ['merchant',"category",'gender','city','state','job']
encode = LabelEncoder()
for col in cat_col:
    train_df[col] = encode.fit_transform(train_df[col])
    test_df[col] = encode.fit_transform(test_df[col])
```

```python
test_df
```

MODEL TRAINING & DATA SPLITIING 

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from imblearn.over_sampling import SMOTE
```

```python
X_train = train_df.drop(columns=['is_fraud'])
y_train = train_df['is_fraud']

X_test = test_df.drop(columns=['is_fraud'])
y_test = test_df['is_fraud']
```

```python
#using SMOTE
smote = SMOTE(random_state=42)

X_train, y_train = smote.fit_resample(X_train, y_train)
```

```python
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model1 = RandomForestClassifier(n_estimators=100, random_state=42)
model2 = model = xgb.XGBClassifier(n_estimators=100, learning_rate=0.1, )

model1.fit(X_train, y_train)
model2.fit(X_train, y_train)
```

MODEL PREDICTION

```python
y_pred1 = model1.predict(X_test)

y_pred_prob = model2.predict_proba(X_test)[:,1]
y_pred2 = (y_pred_prob > 0.3).astype(int)
```

MODEL1 EVALUATION

```python
print(confusion_matrix(y_test, y_pred1))
print(classification_report(y_test, y_pred1))
print("ROC AUC:", roc_auc_score(y_test, y_pred1))
```

MODEL2 EVALUATION

```python
print(confusion_matrix(y_test, y_pred2))
print(classification_report(y_test, y_pred2))

print("ROC AUC:", roc_auc_score(y_test, y_pred_prob))
```
